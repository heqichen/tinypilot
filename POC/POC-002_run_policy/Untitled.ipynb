```python
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import numpy as np
import time

for i in range(20):
    clear_output(wait=True)
    fig, axs = plt.subplots(1, 2, figsize=(10, 4))
    
    x = np.arange(10)
    y1 = np.sin(x + i * 0.2)
    y2 = np.cos(x + i * 0.2)
    
    axs[0].plot(x, y1, color='blue')
    axs[0].set_title('sin')
    axs[1].plot(x, y2, color='red')
    axs[1].set_title('cos')
    
    plt.tight_layout()
    display(fig)
    plt.close(fig)
    time.sleep(0.2)
```
