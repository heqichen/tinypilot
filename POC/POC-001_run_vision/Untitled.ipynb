```python
from convert_images import crop_and_resize_image, display_image, convert_to_yuv420, display_grayscale_image, yuv420_to_rgb, display_rgb_image, loadys, loaduv
import numpy as np
```

```python
resized_image = crop_and_resize_image('res/1.jpg')
display_image(resized_image)
Y, U, V = convert_to_yuv420(resized_image)
```

```python
Y, U, V
```

```python
display_grayscale_image(Y, 512, 256)
display_grayscale_image(U, 256, 128)
display_grayscale_image(V, 256, 128)
```

```python
WIDTH = 512
HEIGHT = 256
buffer_size = WIDTH * HEIGHT * 3 // 2
out_frame = np.zeros(buffer_size, dtype=np.uint8)
loadys(Y, out_frame, WIDTH, HEIGHT)
loaduv(U, out_frame, WIDTH*HEIGHT, WIDTH, HEIGHT)
loaduv(V, out_frame, WIDTH*HEIGHT+WIDTH*HEIGHT//4, WIDTH, HEIGHT)
display_grayscale_image(out_frame, WIDTH, HEIGHT)
```

```python


rgb = yuv420_to_rgb(Y, U, V, 512, 256)
```

```python

display_rgb_image(rgb, 512, 256)
```
